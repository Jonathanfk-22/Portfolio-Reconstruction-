```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('datathon_train.csv', parse_dates=['Date'])
df.head()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

# Top: VEGA stock price
ax1.plot(df['Date'], df['VEGA'], color='steelblue', linewidth=1.5, label='VEGA')
ax1.set_ylabel('Price')
ax1.set_title('VEGA Stock Price')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Bottom: Call and Put options
ax2.plot(df['Date'], df['VEGA_C130'], color='green', linewidth=1.5, label='Call (Strike 130)')
ax2.plot(df['Date'], df['VEGA_P110'], color='red', linewidth=1.5, label='Put (Strike 110)')
ax2.set_ylabel('Option Price')
ax2.set_title('VEGA Options')
ax2.set_xlabel('Date')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
